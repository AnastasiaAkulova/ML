Так как датасет boston был удален из библиотеки по причине "расизма", пришлось находить его содержимое самостоятельно и записывать в csv таблицу.

```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
data = pd.read_csv('load_boston.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
TARGET_KEY = 'MEDV'
X_data, y_data = data.drop(columns=[TARGET_KEY]), data[[TARGET_KEY]]
X_test, X_train, y_test, y_train = train_test_split(X_data, y_data)
```

```python
svr = SVR(kernel='linear')
svr.fit(X_train, y_train)
svr_pred = svr.predict(X_test)
svr_pred
```

```python
gb = GradientBoostingRegressor()
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)
gb_pred
```

Для сравнения качества моделей используем метрики средней абсолютной ошибки и коэффициента детерминации. Первый покажет, насколько предсказанные значения отклоняются от настоящих. Второй покажет, насколько точно модель предсказывает значения в целом.

```python
svr_mae = mean_absolute_error(y_test, svr_pred)
gb_mae = mean_absolute_error(y_test, gb_pred)
```

```python
svr_r2 = r2_score(y_test, svr_pred)
gb_r2 = r2_score(y_test, gb_pred)
```

```python
pd.DataFrame.from_dict({
    'SVR': {
        'MAE': round(svr_mae, 5),
        'R2': f'{round(svr_r2 * 100, 3)}%'
    },
    'GB': {
        'MAE': round(gb_mae, 5),
        'R2': f'{round(gb_r2 * 100, 3)}%'
    }
})
```

По результатам можно сказать, что модели можно признать приемлемыми. Также можно заметить, что градиентный бустинг справляется с задачей намного лучше, выдавая хорошие результаты.
